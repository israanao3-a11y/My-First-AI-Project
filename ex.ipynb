```python

%pip install matplotlib 
import matplotlib.pyplot as plt

# 1. البيانات (الدراسة بالذكاء الاصطناعي يحتاج بيانات!)
ساعات_الدراسة = [1, 3, 4, 6, 8]  # (أقصر، أقصر، وسط، أطول، أطول)
الدرجة_النهائية = [12, 14, 16, 18, 20] # (لاحظي كيف تزيد الدرجة مع زيادة الدراسة)

# 2. أمر الرسم (نرسم نقاط صغيرة لكل طالب)
plt.scatter(ساعات_الدراسة, الدرجة_النهائية, color='blue', label='الطلاب')

# 3. تزيين الرسم (إضافة العناوين)
plt.title('العلاقة بين ساعات الدراسة والدرجة النهائية (مثال لبيانات الـ AI)')
plt.xlabel('ساعات الدراسة')
plt.ylabel('الدرجة النهائية')

# 4. عرض الرسم (نفس اللحظة!)
plt.show()
```
